```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc-house-data.csv')
df.info()
```

2.1 Изучите распределение домов от наличия вида на набережную

```python
df['count']= 1
df.head()
```

```python
df['waterfront'].value_counts(normalize=True)*100
```

Вывод: только 0,75% домов имеют вид на набережную.

2.2 Изучите распределение этажей домов

```python
round(df['floors'].value_counts(normalize=True)*100)
```

```python
plt.figure()
sns.histplot(df['floors'], bins = 15)
plt.title('Распределение домов по этажности')
plt.xlabel('Кол-во этажей')
plt.ylabel('Количество домов')
```

Вывод: 49% домов  имеют 1 этаж, 9% - один этаж с мансардой, 38% - 2 этажа, 1%  2 этажа с мансардой, 3% - 3 этажа. 
Доля более высоких этажей близка к нулю.

2.3 Изучите распределение состояния домов.
 Постройте график
- Сделайте выводы

```python
round(df['condition'].value_counts(normalize=True)*100)
```

```python
plt.figure()
plt.hist(df['condition'], bins = 15)
plt.title('Распределение домов по состоянию')
plt.xlabel('Состояние')
plt.ylabel('Количество домов');
```

```python
cond_group = df.groupby('condition')['count'].agg(count='count').reset_index()
cond_group
```

```python
plt.figure()
plt.plot(cond_group['condition'], cond_group['count'])
plt.title('Распределение домов по состоянию')
plt.xlabel('состояние')
plt.ylabel('Количество');
```

Вывод: Более 65% домов находятся в удовлетворительном, 26% - в хорошем, 8% - в отличном состоянии и только 9% домов имеют неудовлетворительное и плохое состояние.
