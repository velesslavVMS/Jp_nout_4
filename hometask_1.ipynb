1.4 Изучите распределение года постройки
 Постройте график
- Назовите график
- Сделайте именование оси x и оси y
- Сделайте выводы

```python
df['count']= 1
df.head()
```

```python
yrgroup = df.groupby('yr_built')['count'].agg(count='count').reset_index()
yrgroup
```

```python
plt.figure()
plt.plot(yrgroup['yr_built'], yrgroup['count'])
plt.title('Распределение домов по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
```

Вывод: количество построенных домов имеет явно выраженную тенденцию к росту. С большой вероятностью может быть аппроксимировано линейной зависимостью.

```python
plt.figure()
sns.lineplot(x = yrgroup['yr_built'], y = yrgroup['count'])
plt.title('Распределение домов по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество')
```
